```python
import pickle
import pandas as pd
```

```python
sentFile = pickle.load(open("/scratch/anamikas/hyp/postVac/jan30/textOfArticles/train/byArticle/wordSentLen.p", "rb"))
truthFile = pickle.load(open("/scratch/anamikas/hyp/v2/dataset/groundTruth/train/byArticle/groundTruthbyArticle.p", "rb"))
```

```python
sentFile1 = pickle.load(open("/scratch/anamikas/hyp/postVac/jan30/textOfArticles/train/byPublisher/wordSentLen.p", "rb"))
truthFile1 = pickle.load(open("/scratch/anamikas/hyp/v2/dataset/groundTruth/train/byPublisher/groundTruthbyPublisher.p", "rb"))
```

```python
NoOfSenttruth = {}
NoOfSenttruth['true'] = []
NoOfSenttruth['false'] = []
```

```python
for key, vals in sentFile.items():
    if truthFile[key]['hyperpartisan'] == "true":
        NoOfSenttruth['true'].append(sentFile[key]['sentenceLen'])
    else:
        NoOfSenttruth['false'].append(sentFile[key]['sentenceLen'])
```

```python
for key, vals in sentFile1.items():
    if truthFile1[key]['hyperpartisan'] == "true":
        NoOfSenttruth['true'].append(sentFile1[key]['sentenceLen'])
    else:
        NoOfSenttruth['false'].append(sentFile1[key]['sentenceLen'])
```
