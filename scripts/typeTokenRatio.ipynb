'''
Finding type token ratio on the articles text.
id : [TTR1, TTR2, TTR3......................avgTTR]
'''

```python
dataPath = "/scratch/anamikas/hyp/postVac/jan30/textOfArticles/train/byArticle/articletext.p"
```

```python
import nltk
import pickle
import string
articleText = pickle.load(open(dataPath, "rb"))
```

```python
exclude = set(string.punctuation)
```

```python
result = {}
for id, text in articleText.items():
    result[id] = {}
    result[id]['length'] = []
    result[id]['avg'] = 0
    sents = nltk.sent_tokenize(text)
    for sent in sents:
        sent = " ".join(word.lower() for word in sent.split())
        sent = ''.join(ch for ch in sent if ch not in exclude)
        tokenised_text = nltk.word_tokenize(sent)
        if len(tokenised_text)==0:
            continue
        uniqueWords = {}
        for words in tokenised_text:
            if words not in uniqueWords:
                uniqueWords[words] = 1
        result[id]['length'].append(len(uniqueWords)/len(tokenised_text))
    result[id]['avg'] = (sum(result[id]['length'])/len(result[id]['length']))
```

```python
for id, lst in result.items():
    print(id, result[id]['avg'])
```

```python
results = {}
for id, text in articleText.items():
    results[id] = {}
    results[id]['length'] = []
    results[id]['avg'] = 0
    sents = nltk.sent_tokenize(text)
    for sent in sents:
        sent = " ".join(word.lower() for word in sent.split())
        sent = ''.join(ch for ch in sent if ch not in exclude)
        tokenised_text = nltk.word_tokenize(sent)
        if len(tokenised_text)==0:
            continue
        results[id]['length'].append(len(set(tokenised_text))/len(tokenised_text))
    results[id]['avg'] = (sum(results[id]['length'])/len(results[id]['length']))
```
